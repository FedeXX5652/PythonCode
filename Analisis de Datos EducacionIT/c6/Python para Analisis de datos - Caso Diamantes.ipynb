## Boxplot

Tambin conocido como diagrama de caja y bigote. Es un mtodo estandarizado para representar grficamente una serie de datos numricos a travs de sus cuartiles. De esta manera, el diagrama de caja muestra a simple vista la mediana y los cuartiles de los datos, pudiendo tambin representar los valores atpicos de estos(outliers).

Recordas que deca que el histograma es muy til para observar cmo se distribuye tu variable numrica? El Boxplot tambin lo es. Si lo miras con mas detalle, incluso te revelar ms informacin en menos tiempo.

El mximo inters del box-plot es visualizar la distribucin de una variable numrica de la manera ms simplificada posible. Slo utiliza los valores de los cuartiles, los extremos (el mnimo y el mximo) y valores raros o outliers.

No depende de valores ponderados como la media. Simplemente se fija en las caractersticas de la posicin. En el fondo, primero ordena la variable y despus la caracteriza segn la posicin de los cuartiles y los valores extremos.

La gracia del boxplot es visualizar una variable numrica de diferentes grupos o categoras. Te facilita la vida cuando comparas varios grupos.

```python
# BOX Carat / Claridad
diamonds.boxplot(column="carat", by ="clarity", figsize=(8,8), notch = True)
plt.show()

# notch nos permite hacer una muesca en la mediana para marcar donde se ubica dicho valor.
```

La mayoria de los diamantes se situan en la parte inferior de la escala
Muchos outliers (mas que 3.5, ya que el maximo 1.5 del rango intercuartilico es cerca de 2 y a partir de ahi toma como outliers

```python
# Precio / Cut
green_diamond = dict(markerfacecolor="g", marker="D")
diamonds.boxplot(column="price", by ="cut", figsize=(8,8), flierprops=green_diamond )
plt.show()

# Creamos un diccionario con las propiedades de los outliers para configurarlos con el color y el tipo de marcador
```

```python
# Precio / Color
diamonds.boxplot(column="price", by ="color", figsize=(8,8), showfliers=False, vert=False)
plt.show()

# Boxplot sin outliers
# Boxplot horizontal
```

## Densidades

## Tabla de frecuencias y barplot

```python
carat_table = pd.crosstab(index=diamonds['clarity'],
                          columns='count')
print(carat_table)
```

```python
carat_table.plot(kind="bar", figsize=(8,8), alpha=0.4, grid=True)
plt.show()
```

```python
carat_table2=pd.crosstab(index=diamonds["clarity"], columns=diamonds['color'])
print(carat_table2)
carat_table2.plot(kind="bar", figsize=(8,8), alpha=0.9, grid=True, stacked=True)
plt.show()
```

```python
carat_table2=pd.crosstab(index=diamonds["color"], columns=diamonds['cut'])
print(carat_table2)
carat_table2.plot(kind="bar", figsize=(8,8), alpha=0.9, grid=True, stacked=False)
plt.show()
```
